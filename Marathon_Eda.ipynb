# What is object datatype in Python?


In pandas, object dtype is used to represent any kind of Python object, but most commonly:

a)Text/string values

b)Mixed types in a column (e.g., numbers + strings)

c)Custom Python objects

Note:
object dtype is flexible but less memory-efficient than native types (like int64, float64, or string).

For text data, pandas is moving toward using string dtype instead of object.

Always good to check and convert object columns when doing numeric operations.

This column holds generic Python objects — most likely strings, but could be anything.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=df3, x='athlete_gender' ,  hue="athlete_gender")
plt.title('Gender Distribution')
plt.show()
```

```python
sns.barplot(data=df3, x='athlete_gender', y='athlete_average_speed', estimator='mean')
plt.title('Average Speed by Gender')

plt.show()
```

```python
sns.scatterplot(data=df3, x='athlete_age', y='athlete_average_speed', hue='athlete_gender')
plt.title('Variation of Average Speed with Age')
plt.ylim(0, 25)
plt.show()
```

```python
We can overlay a regression line or trend line on top of our scatterplot using sns.regplot() 
```

```python
sns.scatterplot(data=df3, x='athlete_age', y='athlete_average_speed', hue='athlete_gender')
sns.regplot(data=df3, x='athlete_age', y='athlete_average_speed', scatter=False, color='black', line_kws={'linewidth': 2})
plt.title('Variation of Average Speed with Age')
plt.ylim(0, 25)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the distribution plot
sns.displot(df3['athlete_average_speed'], kde=True, height=5, aspect=1.5)

# Add vertical line for mean
plt.axvline(df3['athlete_average_speed'].mean(), color='blue', linestyle='--', label='Mean')

# Add vertical line for median
plt.axvline(df3['athlete_average_speed'].median(), color='green', linestyle='-', label='Median')

# Add legend and title
plt.legend()
plt.title('Distribution of Athlete Average Speed')
plt.xlabel('Average Speed')

plt.show()
```
